```python
#-----------------------------------------------------------Gráfico de Performance
#------------------------------------------------------------------Mensal
#-----------------------------------------------------de Acordo c/ Nome do 'Requisitante',
#---------------------------------------------------'Veículo/Equip.' & o Ano da 'Data Req.'
#-------------------------------------------1 Gráfico por Mês. PRESSIONANDO "ENTER" p/ ir p/ Próx Mês

import pandas as pd
import plotly.express as px
#Read CSV arqv
tabela = pd.read_csv("Abastecimento_Caminhao.csv", delimiter=';', on_bad_lines='skip')
#Columns t/exclude
colunas_para_excluir = ["Requisição", "Horim. Equip.", "Vlr. Unitário"]
colunas_existentes = [col for col in colunas_para_excluir if col in tabela.columns]
if colunas_existentes:
    tabela = tabela.drop(columns=colunas_existentes)
else:
    print("Nenhuma das colunas especificadas foi encontrada no DataFrame.")
#Removing NaNs
tabela = tabela.dropna()
#Treating columns'Litros'|'Km Rodados'
tabela['Litros'] = tabela['Litros'].astype(str).str.replace(',', '').str[:-2]
tabela['Vlr. Total'] = tabela['Vlr. Total'].astype(str).str.replace(',', '')#.str[:-2]
tabela['Km Rodados'] = pd.to_numeric(tabela['Km Rodados'], errors='coerce')
tabela['Litros'] = pd.to_numeric(tabela['Litros'], errors='coerce')
tabela['Vlr. Total'] = pd.to_numeric(tabela['Vlr. Total'], errors='coerce')
tabela['Km Atual'] = pd.to_numeric(tabela['Km Atual'], errors='coerce')
#print(tabela.dtypes)
tabela = tabela.dropna(subset=['Km Rodados', 'Litros', 'Vlr. Total', 'Km Atual', 'Veículo/Equip.', 'Requisitante'])
#Solicit requisitante'sname
nameReq = input("Requisitante (exemplo: 'GESSE'): ")
#Functn t/filtr data using-requisitante'sname
def name_Requ_value(nameReq, tabela):
    driverName = nameReq
    #Verify if  requisitante'name->column 'Requisitante'
    if tabela['Requisitante'].str.contains(driverName, case=False, na=False).any():
        consultReq = tabela[tabela['Requisitante'].str.contains(driverName, case=False, na=False)]
        #Show t-drivr veicls
        veiculos_unicos = consultReq['Veículo/Equip.'].unique()
        print("Veículos do requisitante:", veiculos_unicos)
        #Solicit veícls selection
        selected_vehicles = input("Digite os veículos a serem selecionados (separados por vírgulas): ").split(',')
        selected_vehicles = [vehicle.strip() for vehicle in selected_vehicles]
        #Filtr tbl accrding t/t-veicls selectd
        filtered_table = consultReq[consultReq['Veículo/Equip.'].isin(selected_vehicles)]
        ano = input("Digite o ano (AAAA): ")
        #Filtr tbl by year givn
        filtered_table['Data Req.'] = pd.to_datetime(filtered_table['Data Req.'], format='%d/%m/%Y', errors='coerce')
        filtered_table = filtered_table.dropna(subset=['Data Req.'])
        filtered_table['Ano'] = filtered_table['Data Req.'].dt.year
        filtered_table['Mês'] = filtered_table['Data Req.'].dt.month
        #Filtr by t-year given
        filtered_table = filtered_table[filtered_table['Ano'] == int(ano)]
        #Make sure t-column be in-corct-formt
        filtered_table['Litros'] = pd.to_numeric(filtered_table['Litros'], errors='coerce')
        filtered_table['Km Rodados'] = pd.to_numeric(filtered_table['Km Rodados'], errors='coerce')
        #Group By Month/Calc SUM->values needed
        resumo_mensal = filtered_table.groupby('Mês').agg({
            'Litros': 'sum',
            'Km Rodados': 'sum'
        }).reset_index()
        #Calc Km per Litr
        resumo_mensal['Km por Litro'] = resumo_mensal['Km Rodados'] / resumo_mensal['Litros']
        #Creating montly grafcs
        for _, row in resumo_mensal.iterrows():
            mes = row['Mês']
            fig = px.histogram(resumo_mensal[resumo_mensal['Mês'] == mes], x='Mês', y='Km por Litro',
                         title=f"Desempenho Mensal de {driverName} - Mês {mes}",
                         labels={'Km por Litro': 'Km por Litro', 'Mês': 'Mês'})
            
            fig.update_layout(xaxis_title="Mês", yaxis_title="Km por Litro")
            
            fig.show()
            
            input("Pressione Enter para ver o próximo gráfico...")
    else:
        print("Não há 'Requisitante' com esse Nome!")
#Call functn c/name fornecd tbl
name_Requ_value(nameReq, tabela)
```

```python
#------------------------------------------------------------ Comparar Performance do Motoristas('Requisitante')
#----------------------------------------------------------------- C/um Motorista de Melhor Performance C/
#-------------------------------------------------------------------- O Mesmo Caminhão('Veículo/Equip.')

import pandas as pd
import plotly.express as px
tabela = pd.read_csv("Abastecimento_Caminhao.csv", delimiter=';', on_bad_lines='skip')
tabela['Vlr. Total'] = pd.to_numeric(tabela['Vlr. Total'].astype(str).str.replace(',', ''), errors='coerce')
tabela['Km Rodados'] = pd.to_numeric(tabela['Km Rodados'], errors='coerce')
tabela['Litros'] = pd.to_numeric(tabela['Litros'], errors='coerce')

nameReq = input("Requisitante (exemplo: 'GESSE'): ")
def name_Requ_value(nameReq, tabela):
    driverName = nameReq
    
    if tabela['Requisitante'].str.contains(driverName, case=False, na=False).any():
        consultReq = tabela[tabela['Requisitante'].str.contains(driverName, case=False, na=False)]
        
        veiculos_unicos = consultReq['Veículo/Equip.'].unique()
        print("Veículos do requisitante:", veiculos_unicos)
        
        selected_vehicles = input("Digite os veículos a serem selecionados (separados por vírgulas): ").split(',')
        selected_vehicles = [vehicle.strip() for vehicle in selected_vehicles]
        
        filtered_table = consultReq[consultReq['Veículo/Equip.'].isin(selected_vehicles)]

        filtered_table['Lucro'] = filtered_table['Vlr. Total'] - filtered_table['Km Rodados']
        filtered_table['Km por Litro'] = filtered_table['Km Rodados'] / filtered_table['Litros']
        
        resumo_selecionado = filtered_table.agg({
            'Lucro': 'sum',
            'Km por Litro': 'mean',
            'Km Rodados': 'sum',
            'Litros': 'sum'
        })

        tabela['Lucro'] = tabela['Vlr. Total'] - tabela['Km Rodados']
        tabela['Km por Litro'] = tabela['Km Rodados'] / tabela['Litros']

        melhor_resultado = tabela.groupby(['Requisitante', 'Veículo/Equip.']).agg({
            'Lucro': 'sum',
            'Km por Litro': 'mean',
            'Km Rodados': 'sum',
            'Litros': 'sum'
        }).idxmax()

        melhores_requisitante, melhores_veiculo = melhor_resultado['Lucro']

        melhor_table = tabela[(tabela['Requisitante'] == melhores_requisitante) & (tabela['Veículo/Equip.'] == melhores_veiculo)]
        resumo_melhor = melhor_table.agg({
            'Lucro': 'sum',
            'Km por Litro': 'mean',
            'Km Rodados': 'sum',
            'Litros': 'sum'
        })

        comparacao = pd.DataFrame({
            'Métrica': ['Lucro', 'Km por Litro', 'Km Rodados', 'Litros'],
            f'{driverName} - {selected_vehicles[0]}': resumo_selecionado.values,
            f'Melhor - {melhores_requisitante} - {melhores_veiculo}': resumo_melhor.values
        })

        fig = px.bar(comparacao, x='Métrica', y=comparacao.columns[1:],
                     title='Comparação de Métricas',
                     barmode='group',
                     labels={'value': 'Valor', 'variable': 'Legenda'})
        
        fig.update_layout(xaxis_title='Métrica', yaxis_title='Valor')
        
        fig.show()        
    else:
        print("Não há 'Requisitante' com esse Nome!")

name_Requ_value(nameReq, tabela)
```

```python
#------------------------------------------------------------ Comparar Performance do Motoristas('Requisitante')
#----------------------------------------------------------------- C/um Motorista de Melhor Performance C/
#-------------------------------------------------------------------- O Mesmo Caminhão('Veículo/Equip.')

import pandas as pd
import plotly.express as px
#Arqv CSV
tabela = pd.read_csv("Abastecimento_Caminhao.csv", delimiter=';', on_bad_lines='skip')
#Conversão coluns p/numéric
tabela['Vlr. Total'] = pd.to_numeric(tabela['Vlr. Total'].astype(str).str.replace(',', ''), errors='coerce')
tabela['Km Rodados'] = pd.to_numeric(tabela['Km Rodados'], errors='coerce')
tabela['Litros'] = pd.to_numeric(tabela['Litros'], errors='coerce')
#Solicit nome d-requisitnt
nameReq = input("Requisitante (exemplo: 'GESSE'): ")
#Functn p/filtr dados byname d-requisitnt/veícl
def name_Requ_value(nameReq, tabela):
    driverName = nameReq
    #Verify if name-requistnt tá in-column'Requisitante'
    if tabela['Requisitante'].str.contains(driverName, case=False, na=False).any():
        consultReq = tabela[tabela['Requisitante'].str.contains(driverName, case=False, na=False)].copy()
        #Show veícls únics p/Driver
        veiculos_unicos = consultReq['Veículo/Equip.'].unique()
        print("Veículos do requisitante:", veiculos_unicos)
        #Solict seleç veícls
        selected_vehicles = input("Digite os veículos a serem selecionados (separados por vírgulas): ").split(',')
        selected_vehicles = [vehicle.strip() for vehicle in selected_vehicles]
        #Filtr tbl w/veícls selectd
        filtered_table = consultReq[consultReq['Veículo/Equip.'].isin(selected_vehicles)].copy()
        #Calc métrics p/Driver
        filtered_table.loc[:, 'Lucro'] = filtered_table['Vlr. Total'] - filtered_table['Km Rodados']
        filtered_table.loc[:, 'Km por Litro'] = filtered_table['Km Rodados'] / filtered_table['Litros']
        
        resumo_selecionado = filtered_table.agg({
            'Lucro': 'sum',
            'Km por Litro': 'mean',
            'Km Rodados': 'sum',
            'Litros': 'sum'
        })
        #Find Driver w/best results
        tabela.loc[:, 'Lucro'] = tabela['Vlr. Total'] - tabela['Km Rodados']
        tabela.loc[:, 'Km por Litro'] = tabela['Km Rodados'] / tabela['Litros']
        melhor_resultado = tabela.groupby(['Requisitante', 'Veículo/Equip.']).agg({
            'Lucro': 'sum',
            'Km por Litro': 'mean',
            'Km Rodados': 'sum',
            'Litros': 'sum'
        }).idxmax(skipna=True)

        melhores_requisitante, melhores_veiculo = melhor_resultado['Lucro']
        melhor_table = tabela[(tabela['Requisitante'] == melhores_requisitante) & (tabela['Veículo/Equip.'] == melhores_veiculo)].copy()
        resumo_melhor = melhor_table.agg({
            'Lucro': 'sum',
            'Km por Litro': 'mean',
            'Km Rodados': 'sum',
            'Litros': 'sum'
        })
        #Compare-métrics
        comparacao = pd.DataFrame({
            'Métrica': ['Lucro', 'Km por Litro', 'Km Rodados', 'Litros'],
            f'{driverName} - {selected_vehicles[0]}': resumo_selecionado.values,
            f'Melhor - {melhores_requisitante} - {melhores_veiculo}': resumo_melhor.values
        })

        #Make t-comparativ-gráfc
        fig = px.bar(comparacao, x='Métrica', y=comparacao.columns[1:],
                     title='Comparação de Métricas',
                     barmode='group',
                     labels={'value': 'Valor', 'variable': 'Legenda'})
        
        fig.update_layout(xaxis_title='Métrica', yaxis_title='Valor')
        
        fig.show()
    
    else:
        print("Não há 'Requisitante' com esse Nome!")
#Call t-Funçtn c/Name+table
name_Requ_value(nameReq, tabela)
```
