```python
import pandas as pd
import plotly.express as px
tabela = pd.read_csv("Abastecimento_Caminhao.csv", delimiter=';', on_bad_lines='skip')
colunas_para_excluir = ["Requisição", "Hora Abast.", "Obs.", "Abast. Externo", "Data Req.", "Combustível", "Horim. Equip.", "Vlr. Unitário"]
colunas_existentes = [col for col in colunas_para_excluir if col in tabela.columns]
if colunas_existentes:
    tabela = tabela.drop(columns=colunas_existentes)
else:
    print("Nenhuma das colunas especificadas foi encontrada no DataFrame.")

tabela = tabela.dropna()
#display(tabela)

tabela['Litros'] = tabela['Litros'].astype(str).str.replace(',', '').str[:-2]
tabela['Vlr. Total'] = tabela['Vlr. Total'].astype(str).str.replace(',', '')#.str[:-2]
tabela['Km Rodados'] = pd.to_numeric(tabela['Km Rodados'], errors='coerce')
tabela['Litros'] = pd.to_numeric(tabela['Litros'], errors='coerce')
tabela['Vlr. Total'] = pd.to_numeric(tabela['Vlr. Total'], errors='coerce')
tabela['Km Atual'] = pd.to_numeric(tabela['Km Atual'], errors='coerce')
tabela = tabela.dropna(subset=['Km Rodados', 'Litros', 'Vlr. Total', 'Km Atual', 'Veículo/Equip.', 'Requisitante'])
agrupado = tabela.groupby(['Veículo/Equip.', 'Requisitante']).agg({
    'Litros': 'sum',
    'Vlr. Total': 'sum',
    'Km Rodados': 'sum',
    'Km Atual': 'max'
}).reset_index()
#print("DataFrame agrupado:")
#display(agrupado)
agrupado['Km por Litro'] = agrupado['Km Rodados'] / agrupado['Litros']

agrupado['Litros'] = agrupado['Litros'].apply(lambda x: '{:,.2f}'.format(x))
agrupado['Vlr. Total'] = agrupado['Vlr. Total'].apply(lambda x: '{:,.2f}'.format(x))
#print("DataFrame agrupado com formatação de vírgulas:")
#display(agrupado)
for coluna in tabela['Vlr. Total']:
    grafico = px.histogram(tabela, x='Vlr. Total', color="Veículo/Equip.")

    grafico.show()
```

```python
#---------------------------------------------Top10 Veículos/Equip com Mais/Menos LUCRO(km Rodados - Vlr. Total)

import pandas as pd
import plotly.express as px
tabela = pd.read_csv("Abastecimento_Caminhao.csv", delimiter=';', on_bad_lines='skip')
colunas_para_excluir = ["Requisição", "Hora Abast.", "Obs.", "Abast. Externo", "Data Req.", "Combustível", "Horim. Equip.", "Vlr. Unitário"]
colunas_existentes = [col for col in colunas_para_excluir if col in tabela.columns]
if colunas_existentes:
    tabela = tabela.drop(columns=colunas_existentes)
else:
    print("Nenhuma das colunas especificadas foi encontrada no DataFrame.")
tabela = tabela.dropna()
tabela['Litros'] = tabela['Litros'].astype(str).str.replace(',', '').str[:-2]
tabela['Vlr. Total'] = tabela['Vlr. Total'].astype(str).str.replace(',', '')#.str[:-2]
tabela['Km Rodados'] = pd.to_numeric(tabela['Km Rodados'], errors='coerce')
tabela['Litros'] = pd.to_numeric(tabela['Litros'], errors='coerce')
tabela['Vlr. Total'] = pd.to_numeric(tabela['Vlr. Total'], errors='coerce')
tabela['Km Atual'] = pd.to_numeric(tabela['Km Atual'], errors='coerce')
#print(tabela.dtypes)
tabela = tabela.dropna(subset=['Km Rodados', 'Litros', 'Vlr. Total', 'Km Atual', 'Veículo/Equip.', 'Requisitante'])
agrupado = tabela.groupby(['Veículo/Equip.', 'Requisitante']).agg({
    'Litros': 'sum',
    'Vlr. Total': 'sum',
    'Km Rodados': 'sum',
    'Km Atual': 'max'
}).reset_index()
#print(agrupado.dtypes)
agrupado['Lucro'] = agrupado['Km Rodados'] - agrupado['Vlr. Total']
top10_lucro = agrupado.nlargest(10, 'Lucro')
bottom10_lucro = agrupado.nsmallest(10, 'Lucro')
top_bottom_lucro = pd.concat([top10_lucro, bottom10_lucro])

grafico = px.bar(top_bottom_lucro, x='Veículo/Equip.', y='Lucro', color='Lucro',
             color_continuous_scale=[(0, 'blue'), (0.5, 'blue'), (0.5, 'red'), (1, 'red')],
             hover_data=['Requisitante'])

grafico.update_layout(title="Top 10 e Bottom 10 Veículos/Equipamentos por Lucro",
                  xaxis_title="Veículo/Equip.",
                  yaxis_title="Lucro (Km Rodados - Vlr. Total)",
                  xaxis_tickangle=-45)

grafico.show()
```

```python
#--------------------------------------------------------------------VEÍCULOS c/
#-----------------------------------------------------------------MAIORES/MENORES
#---------------------------------------------------------KM por LITRO(Km Rodados / Litros)

import pandas as pd
import plotly.express as px
#---------Leitura arqv CSV
tabela = pd.read_csv("Abastecimento_Caminhao.csv", delimiter=';', on_bad_lines='skip')
#---------Colunas p/ exclr
colunas_para_excluir = ["Requisição", "Hora Abast.", "Obs.", "Abast. Externo", "Data Req.", "Combustível", "Horim. Equip.", "Vlr. Unitário"]
colunas_existentes = [col for col in colunas_para_excluir if col in tabela.columns]
if colunas_existentes:
    tabela = tabela.drop(columns=colunas_existentes)
else:
    print("Nenhuma das colunas especificadas foi encontrada no DataFrame.")
#----------Remov NaNs
tabela = tabela.dropna()
#----------Tratamento-> colunas 'Litros'/'Km Rodados'
tabela['Litros'] = tabela['Litros'].astype(str).str.replace(',', '').str[:-2]
tabela['Vlr. Total'] = tabela['Vlr. Total'].astype(str).str.replace(',', '')#.str[:-2]
tabela['Km Rodados'] = pd.to_numeric(tabela['Km Rodados'], errors='coerce')
tabela['Litros'] = pd.to_numeric(tabela['Litros'], errors='coerce')
tabela['Vlr. Total'] = pd.to_numeric(tabela['Vlr. Total'], errors='coerce')
tabela['Km Atual'] = pd.to_numeric(tabela['Km Atual'], errors='coerce')
#-----------Remover de NaNs
tabela = tabela.dropna(subset=['Km Rodados', 'Litros', 'Vlr. Total', 'Km Atual', 'Veículo/Equip.', 'Requisitante'])
#-----------Excluir veículos indesejados
valores_para_excluir = ['GERADOR']# ,'9I10', '1629', '7888', '8C39'
tabela = tabela[~tabela['Veículo/Equip.'].isin(valores_para_excluir)]
#------------Agrupar por 'Veículo/Equip.'/'Requisitante'
agrupado = tabela.groupby(['Veículo/Equip.', 'Requisitante']).agg({
    'Litros': 'sum',
    'Vlr. Total': 'sum',
    'Km Rodados': 'sum',
    'Km Atual': 'max'
}).reset_index()
#--------------Calc 'Km por Litro'/arredondar
agrupado['Km por Litro'] = (agrupado['Km Rodados'] / agrupado['Litros']).round(7)
#--------------Seleção-> Top e Bottom
top5 = agrupado.nlargest(5, 'Km por Litro')
bottom23 = agrupado.nsmallest(23, 'Km por Litro')
#--------------Gráfico-> Top
fig_top5 = px.bar(top5, x='Veículo/Equip.', y='Km por Litro', color='Km por Litro',
             hover_data=['Requisitante'])
fig_top5.update_layout(title="Veículos/Equipamentos com MAIOR Km por Litro",
                  xaxis_title="Veículo/Equip.",
                  yaxis_title="Km por Litro",
                  xaxis_tickangle=-45)
#--------------Gráfico-> Bottom
fig_bottom23 = px.bar(bottom23, x='Veículo/Equip.', y='Km por Litro', color='Km por Litro',
             hover_data=['Requisitante'])
fig_bottom23.update_layout(title="Veículos/Equipamentos com MENOR Km por Litro",
                  xaxis_title="Veículo/Equip.",
                  yaxis_title="Km por Litro",
                  xaxis_tickangle=-45)
#--------------Exibir
fig_top5.show()
fig_bottom23.show()
```
